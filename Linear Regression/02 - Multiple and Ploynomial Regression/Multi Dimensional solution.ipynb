# Multi Dimensional solution

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = []
Y = []
for line in open('C:/Users/dubek/Downloads/New folder/data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([float(x1), float(x2), 1]) # add the bias term
    Y.append(float(y))
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
plt.show()
```

apply the equations we learned to calculate a and b.

numpy has a special method for solving Ax = b

so we don't use x = inv(A)*b

note: the * operator does element-by-element multiplication in numpy

np.dot() does what we expect for matrix multiplication

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
Yhat = np.dot(X, w)
```

```python
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print("the r-squared is:", r2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], Y)
ax.plot(sorted(X[:,0]), sorted(X[:,1]), sorted(Yhat))
plt.show()
```
